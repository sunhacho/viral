```python
# í…ŒìŠ¤íŠ¸ ë°ì´í„°ì— ëŒ€í•´ ì˜ˆì¸¡ ìˆ˜í–‰ (í‚¤ì›Œë“œ ê¸°ë°˜ ë¶„ë¥˜ ì œê±°)
def predict_without_keywords(examples):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # ëª¨ë¸ê³¼ ë™ì¼í•œ ì¥ì¹˜ ì„¤ì •
    model.to(device)  # ëª¨ë¸ì„ í•´ë‹¹ ì¥ì¹˜ë¡œ ì´ë™

    predictions = []
    for tweet in examples["ì „ì²˜ë¦¬ëœ íŠ¸ìœ—"]:
        # ëª¨ë¸ì„ ì‚¬ìš©í•œ ì˜ˆì¸¡
        inputs = tokenizer(tweet, return_tensors='pt', padding=True, truncation=True, max_length=512)
        inputs = {k: v.to(device) for k, v in inputs.items()}  # ì…ë ¥ í…ì„œë¥¼ ëª¨ë¸ê³¼ ë™ì¼í•œ ë””ë°”ì´ìŠ¤ë¡œ ì´ë™
        outputs = model(**inputs)
        predictions.append(outputs.logits.argmax(dim=1).item())  # ì˜ˆì¸¡ëœ ë¼ë²¨ ì¶”ê°€

    return predictions

# í‚¤ì›Œë“œ ê°•ì œ ë¶„ë¥˜ë¥¼ ì‚¬ìš©í•˜ì§€ ì•Šê³  ì˜ˆì¸¡
test_predictions = predict_without_keywords(test_data)
test_data = pd.DataFrame(test_data)
test_data['ì˜ˆì¸¡'] = test_predictions

# ê²°ê³¼ í™•ì¸
print(test_data[['ì „ì²˜ë¦¬ëœ íŠ¸ìœ—', 'ë¼ë²¨ë§', 'ì˜ˆì¸¡']])
```

```python
import torch
import torch.nn.functional as F

def predict_proba_without_keywords(examples):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    probabilities = []
    for tweet in examples["ì „ì²˜ë¦¬ëœ íŠ¸ìœ—"]:
        inputs = tokenizer(tweet, return_tensors='pt', padding=True, truncation=True, max_length=512)
        inputs = {k: v.to(device) for k, v in inputs.items()}
        outputs = model(**inputs)

        # ì†Œí”„íŠ¸ë§¥ìŠ¤ë¥¼ í†µí•´ í™•ë¥ ì„ ê³„ì‚°
        probs = F.softmax(outputs.logits, dim=1)
        probabilities.append(probs.detach().cpu().numpy()[0])  # í™•ë¥ ì„ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€

    return probabilities

# ì˜ˆì¸¡ í™•ë¥ ì„ ìƒì„±í•˜ì—¬ ê²°ê³¼ë¥¼ ì €ì¥
koelectra_proba_test = predict_proba_without_keywords(X_test_koelectra)
koelectra_proba_test = np.array(koelectra_proba_test)  # Numpy ë°°ì—´ë¡œ ë³€í™˜
```

```python
# ì˜ˆì¸¡ í™•ë¥  ê³„ì‚°
tabnet_proba_test = clf.predict_proba(X_test_tabnet)

# ì˜ˆì¸¡ í™•ë¥ ì„ Numpy ë°°ì—´ë¡œ ë³€í™˜
tabnet_proba_test = np.array(tabnet_proba_test)
```

```python
# ê° ëª¨ë¸ì— ê°€ì¤‘ì¹˜ ì ìš© (ì˜ˆ: KoELECTRA ë” ì‹ ë¢°)
tabnet_weight = 0.37
koelectra_weight = 0.63

# 1. ê°€ì¤‘ì¹˜ ê¸°ë°˜ í‰ê·  ê³„ì‚°
weighted_voting_proba = (tabnet_weight * tabnet_proba_test + koelectra_weight * koelectra_proba_test)

# 2. ì„ê³„ê°’ì— ë”°ë¼ ìµœì¢… ì˜ˆì¸¡ ìƒì„±
voting_predictions = []  # ìµœì¢… ì˜ˆì¸¡ê°’ì„ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
for weighted_prob in weighted_voting_proba[:, 1]:  # ê°€ì¤‘ í‰ê· ì„ ê¸°ì¤€ìœ¼ë¡œ ì–‘ì„± í´ë˜ìŠ¤ í™•ë¥ ê°’
    if weighted_prob >= 0.365:  # ì„ê³„ê°’ 0.5ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì˜ˆì¸¡
        voting_predictions.append(1)
    else:
        voting_predictions.append(0)

# 3. numpy ë°°ì—´ë¡œ ë³€í™˜
voting_predictions = np.array(voting_predictions)

# ê²°ê³¼ ì¶œë ¥
print("TabNet í™•ë¥ ê°’ (ì–‘ì„± í´ë˜ìŠ¤):", tabnet_proba_test[:, 1])
print("KoELECTRA í™•ë¥ ê°’ (ì–‘ì„± í´ë˜ìŠ¤):", koelectra_proba_test[:, 1])
print("ê°€ì¤‘ì¹˜ ê¸°ë°˜ í‰ê·  í™•ë¥ ê°’:", weighted_voting_proba[:, 1])
print("ìµœì¢… ì˜ˆì¸¡ê°’:", voting_predictions)
```
