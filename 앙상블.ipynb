```python
def tokenize_function(examples):
    return tokenizer(examples['전처리된 트윗'], padding='max_length', truncation=True)

test_data = test_data.map(tokenize_function, batched=True)
```

```python
# 테스트 데이터에 대해 예측 수행 (키워드 기반 분류 제거)
def predict_without_keywords(examples):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # 모델과 동일한 장치 설정
    model.to(device)  # 모델을 해당 장치로 이동

    predictions = []
    for tweet in examples["전처리된 트윗"]:
        # 모델을 사용한 예측
        inputs = tokenizer(tweet, return_tensors='pt', padding=True, truncation=True, max_length=512)
        inputs = {k: v.to(device) for k, v in inputs.items()}  # 입력 텐서를 모델과 동일한 디바이스로 이동
        outputs = model(**inputs)
        predictions.append(outputs.logits.argmax(dim=1).item())  # 예측된 라벨 추가

    return predictions

# 키워드 강제 분류를 사용하지 않고 예측
test_predictions = predict_without_keywords(test_data)
test_data = pd.DataFrame(test_data)
test_data['예측'] = test_predictions

# 결과 확인
print(test_data[['전처리된 트윗', '라벨링', '예측']])
```

```python
import torch
import torch.nn.functional as F

def predict_proba_without_keywords(examples):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    probabilities = []
    for tweet in examples["전처리된 트윗"]:
        inputs = tokenizer(tweet, return_tensors='pt', padding=True, truncation=True, max_length=512)
        inputs = {k: v.to(device) for k, v in inputs.items()}
        outputs = model(**inputs)

        # 소프트맥스를 통해 확률을 계산
        probs = F.softmax(outputs.logits, dim=1)
        probabilities.append(probs.detach().cpu().numpy()[0])  # 확률을 리스트에 추가

    return probabilities

# 예측 확률을 생성하여 결과를 저장
koelectra_proba_test = predict_proba_without_keywords(X_test_koelectra)
koelectra_proba_test = np.array(koelectra_proba_test)  # Numpy 배열로 변환
```

```python
# 예측 확률 계산
tabnet_proba_test = clf.predict_proba(X_test_tabnet)

# 예측 확률을 Numpy 배열로 변환
tabnet_proba_test = np.array(tabnet_proba_test)
```

```python
# 각 모델에 가중치 적용 (예: KoELECTRA 더 신뢰)
tabnet_weight = 0.37
koelectra_weight = 0.63

# 1. 가중치 기반 평균 계산
weighted_voting_proba = (tabnet_weight * tabnet_proba_test + koelectra_weight * koelectra_proba_test)

# 2. 임계값에 따라 최종 예측 생성
voting_predictions = []  # 최종 예측값을 저장할 리스트
for weighted_prob in weighted_voting_proba[:, 1]:  # 가중 평균을 기준으로 양성 클래스 확률값
    if weighted_prob >= 0.365:  # 임계값 0.5를 기준으로 예측
        voting_predictions.append(1)
    else:
        voting_predictions.append(0)

# 3. numpy 배열로 변환
voting_predictions = np.array(voting_predictions)

# 결과 출력
print("TabNet 확률값 (양성 클래스):", tabnet_proba_test[:, 1])
print("KoELECTRA 확률값 (양성 클래스):", koelectra_proba_test[:, 1])
print("가중치 기반 평균 확률값:", weighted_voting_proba[:, 1])
print("최종 예측값:", voting_predictions)
```

```python
true_labels = test_df['라벨링']
```

```python

# 2. 키워드 기반으로 최종 예측값 수정
keywords = ['제품 제공', '광고', '#제품제공', '#광고', '#타채널제품제공', '협찬', "'쿠팡, 파트너'", '제품제공', '#제품지원', '쿠팡파트너', '#대가성광고', "'제품', '제공'", '쿠팡 파트너','#협찬']
  # 무조건 0으로 바꿀 키워드 리스트

# 3. test_df에서 키워드 포함 여부 확인 후 예측값 수정
for i, tweet in enumerate(test_df["전처리된 트윗"]):
    if any(keyword in tweet for keyword in keywords):
        voting_predictions[i] = 0  # 키워드 포함 시 0으로 설정

# 결과 확인
print("최종 수정된 예측값:", voting_predictions)
```
